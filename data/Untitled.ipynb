```python
import os
import pandas as pd 
import glob
```

```python
Path = f'./scraped-data/'

filePath = glob.glob(os.path.join(Path,"*json"))

myList = []


for file in filePath:
    
    myList.append(file)
    salaryDf = pd.concat(
        map(pd.read_json, myList), ignore_index=True)
    
salaryDf = salaryDf.loc[salaryDf['name'] != 'NAME']

salaryDf['salary'] = salaryDf['salary'].str.replace('$','').str.replace(',','').astype(int)

salaryDf= salaryDf.sort_values(['salary'],ascending=False).drop_duplicates(keep='first')

salaryDf
```
